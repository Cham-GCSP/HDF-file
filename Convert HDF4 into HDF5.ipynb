This is the code to convert hdf4 into hdf5.
Data file includes hourly concentration of PM2.5 over 4 years: 2015-2018.
For each day, data is retrieve from 0-7AM.

Outline of code content:
1. Convert hdf4 (h4) to hdf5 (h5)
2. Check and extract value from h5 file


```python
import os

#folder path

dir_path = r'E:\GOCI_PM\PM_NKSK_hdf'

out_h5 = r'E:\GOCI_PM\h5_out_folder'
```

```python
print(dir_path)
print(out_h5)
```

```python
#iterate directory
import subprocess

for path in os.listdir(dir_path):
    #check if current path is a file
    if os.path.isfile(os.path.join(dir_path, path)):
        file_loc = os.path.join(dir_path, path)
        new_file_h5 = os.path.join(out_h5, path.replace('hdf', 'h5'))
    
    try:
        print('processing {}'.format(file_loc))
        string = '"cd C:/Program Files/HDF_Group/H4H5/2.2.5/bin" || h4toh5convert {} {}'.format(file_loc, new_file_h5)
        subprocess.call(string, shell = True)
        print('Done')
        print('==================')
    except:
        continue
```
